# DATA CLEANING
Here we perform missing value treatment. We also
1. First of all we fill **"?"** with np.Nan
2. Check percentage of missing values of variables (columns). If we have a low percentage of missing then some missing values can be replace by the mean or median while other by the mode i.e the most occuring values. If percentages are high then we just remove the columns. 
3. Create a cleaning function to facilitate the cleaning (replacing currency **'$' and ','**  for currency; filling NaN for quantitative an categorical variables). Call the function to clean df_train and df_test. 
4. Checking missing percentages again
5. Treat categorical variables
6. Summary of descriptive statistic after data cleaning
7. visualising data after data cleaning

1. Filling **"?"** using np.Nan

```python
#import numpy as np
df_train=df_train.replace("?",np.NaN)
df_test=df_test.replace("?",np.NaN)
```

2. Check percentage of missing values of variables (columns) for both df_train and df_test

```python
round(100*(df_train.isnull().sum()/len(df_train.index)),2)
```

For df_train, we can see above that missing values are AGE,YOJ,INCOME,HOM_VAL,JOB and CAR_AGE. The highest percentage of missings is 6.45%, which is very small. We can impute all quantitative variables with the median or mean or mode i.e most occuring values. The missing value for categorical variable JOB will be replaced by Unknown 'U' category for further treatment.

```python
round(100*(df_test.isnull().sum()/len(df_test.index)),2)
```

Above, df_test contains 100% missing of TARGET_FLAG (which is normal since it is the target to predict) and TARGET_AMT (which must be removed as predictor from both df_train and df_test)

3. Creating a cleaning function. Notice that ',' is recplace by '.' instead of '' to reduce extremes values.

Get columns names for iteration in the cleaning function

```python
df_train.columns
```

```python
def clean(data):
    data = data.drop(["TARGET_AMT","INDEX"], axis=1) #removing variables that are not pretictors in both df_train, df_test
    columns = ["INCOME","HOME_VAL","BLUEBOOK","OLDCLAIM"] #list of currency variables to be treated
    for col in columns:
        data[col]=data[col].str.replace(',','.') #relacing ',' by '.'
        data[col]=data[col].str.replace("$",'')  #relacing '$' by ''
        data[col]=data[col].astype(float)        #convert currency variables to float
    cols = ["AGE","YOJ","INCOME","HOME_VAL","CAR_AGE"] #list of quantitative variables to be treated
    for col in cols:
        data[col].fillna(data[col].median(), inplace=True) #filling missing value of quantitative variable with median
    data.JOB.fillna("U", inplace=True) #filling missing value of categorical variable 'JOB' with unknown value 'U' 
    return data
```

```python
#Calling the cleaning function of missing for df_train and df_test. We remove the target 'TARGET_FLAG' for the df_test.
df_test = df_test.drop(["TARGET_FLAG"],axis=1)
df_train = clean(df_train)
df_test = clean(df_test)
```

# MODELING AND EVALUATION (MODEL SELECTION)

First of all, our data clearly show that we have a classification problem since we have a predictor and target (client claims, binary dataset). We are in a case of supervised learning and will test 3 appropriates classifier models (Logistic Regression, Random Forest  and SVM).
1. Preparing training data. We split data into train data(to create a model) and validation data (to test accuracy of the model)
2. Models parameters and sampling
3. Initialization of the scores table
4. Looping through samples (e.g. 20 samples) to train and compute the accuracy of each of the 3 models  
5. Create a scores table for the 3 models. 
6. Boxplot to compare the performance of models based on their scores
7. Conclusion on model choice

1. Preparing training data. y is the target and X is the predictor. For X we remove the variable to predict and the index. 

```python
y = df_train["TARGET_FLAG"] #target
X = df_train.drop("TARGET_FLAG", axis=1) #remove variable to predict
```

2. Models parameters and sampling. Let's define **k** the number of sampling (repeating sample depending on the random state) and **s** the number of models to be tested.

```python
k = 20
s = 3
```

3. Initialization of the scores table

```python
cv_accuracy_val = np.zeros(shape=(k,s))
```

4. Looping through samplings

```python
#import warnings
#warnings.filterwarnings("ignore", category=FutureWarning)

for j in range(k) : 
    
    print("\nSampling ->", j+1, ":")
    #from sklearn.model_selection import train_test_split
    # create training and validation dataset for X and y :
    # Split data into train data(to create a model) and validation data (to test accuracy of the model) for each random sample j
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=j)
    
    
    # 1st model : Logistic Regression
    print("Logistic Regression Model:")
    #from sklearn.linear_model import LogisticRegression
    # model definition: LogisticRegression()
    model = LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                                                    intercept_scaling=1, l1_ratio=None, max_iter=200,
                                                    multi_class='auto', n_jobs=None, penalty='l2',
                                                    random_state=None, solver='newton-cg', tol=0.0001, verbose=0,
                                                    warm_start=False)
    # Several solver can be tested solver : {'newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'}, default='lbfgs'
    
    # Training the logistic classifier
    logistic_clf= model.fit(X_train, y_train) 
    # Predict the TARGET_FLAG of X_test
    predictions_logistic = logistic_clf.predict(X_val) 
    #from sklearn.metrics import accuracy_score
    # Accuracy evaluation of the model:accuracy_score()
    accuracy = accuracy_score(predictions_logistic, y_val)
    print("The accuracy of the model is: ", accuracy*100, "%")
    
    
    # 2e model : Random Forest 
    print("Random Forest Model:")
    #from sklearn.ensemble import RandomForestClassifier
    # Model definition : RandomForestClassifier()
    model = RandomForestClassifier(n_estimators=7, verbose=0, max_features=10)
    # Training Random Forest classifier
    randomforest_clf = model.fit(X_train, y_train) 
    # Predict the TARGET_FLAG of X_val
    predictions_randomforest = randomforest_clf.predict(X_val)
    
    # Accuracy evaluation of the model
    accuracy = accuracy_score(predictions_randomforest, y_val)
    print("The accuracy of the model is : ", accuracy*100, "%")


    
    # 3rd model : Support Vector Machine (SVM)
    print("Support Vector Machine (SVM) Model :")
    #from sklearn import svm
    # Model definition : svm.SVC()
    model = svm.SVC(C=1.0, kernel="rbf", degree=3)
    # Training SVM classifier
    svm_clf = model.fit(X_train, y_train) 
    # Predict the TARGET_FLAG of X_val
    predictions_svm =  svm_clf.predict(X_val)
    
    # Accuracy evaluation of the model
    accuracy = accuracy_score(predictions_svm, y_val)
    print("The accuracy of the model is: ", accuracy*100, "%")

    
    # Computing scores of training for each sample (these are scores for the validation test)
    cv_accuracy_val[j,0] = accuracy_score(predictions_logistic, y_val)
    cv_accuracy_val[j,1] = accuracy_score(predictions_randomforest, y_val)
    cv_accuracy_val[j,2] = accuracy_score(predictions_svm, y_val)
    
```

5. Renaming Accuracy Scores Table

```python
cv_accuracy_val = pd.DataFrame(data=cv_accuracy_val, columns=["Logistic", "Random_Forest", "SVM"]) 
cv_accuracy_val #table of scores
```

```python
cv_accuracy_val.mean() #mean of scores
```

6. Boxplots of accuracy scores showing models performances for each of the 3 models

```python
#import matplotlib.pyplot as plt
#%matplotlib inline

boxprops = dict(linestyle='-', linewidth=4, color='k')
medianprops = dict(linestyle='-', linewidth=4, color='k')

plt.figure(figsize=(15,10))
boxplot_accuracy_val = cv_accuracy_val.boxplot(column=["Logistic", "Random_Forest", "SVM"], 
                                                 grid=True,showfliers=False, showmeans=True, 
                                                 boxprops=boxprops, medianprops=medianprops)
plt.ylabel('Accuracy of Prediction/Validation', fontsize=12)
```

7. Conclusion model selection : Based on the training statistics and analysis of boxplot, it is clear that the logistic regression model is more adapted for this classification problem (best mean/median (with 78% accuracy), and reduced interquatile range as compared to the other two models). The logistic regression model can be used in the following step to predict client claims using the data df_test.
